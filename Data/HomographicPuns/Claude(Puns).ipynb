```python
import re
i=0
ClaudeInfer=[]


def extract_numbers_from_text(text):
    # Split the text into lines
    lines = text.split('\n')
    # List to store extracted numbers
    numbers = []

    # Loop through each line
    for line in lines:
        # Check if the line ends with a number, possibly after some text
        match = re.search(r'\b(\d)$', line)
        if match:
            # Extract the number
            number = match.group(1)
            numbers.append(int(number))

    return numbers
fails=[]
for x in all:
    outputs = re.findall(r'[**]*output:[**]*\s*(\d)', x, re.IGNORECASE)
    ls = [int(output) for output in outputs]

    i+=1
    ClaudeInfer+=ls
print(len(ClaudeInfer))
# print(fails)
```
